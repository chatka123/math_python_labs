```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-1,1,10)
y = np.linspace(-1,1,10)
z = lambda x,y: x**2-y**2
x,y = np.meshgrid(x,y)
Z = z(x,y)
List = plt.figure(figsize=(5,5))
ax = plt.axes(projection='3d')
surf = ax.plot_wireframe(x,y,Z,rstride=1,cstride=1)
ax.contourf(x,y,Z,zdir='z',offset=np.min(Z),alpha=0.5)
ax.contourf(x,y,Z,zdir='y',offset=np.min(Z),alpha=0.5)
ax.contourf(x,y,Z,zdir='x',offset=np.min(x),alpha=0.5)
List.colorbar(surf,ax = ax,shrink = 0.5,aspect = 5)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
print('The equation of surface projection to XY:')
print('x**2-y**2=C , where C = const')
print('The equation of surface projection to XZ:')
print('x**2-C=z , where C = const')
print('The equation of surface projection to YZ:')
print('C-y**2=z , where C = const ')
```

```python
import sympy as sp
x,y,z,fi,om,R = sp.symbols('x y z fi om R')
x = R*sp.cos(fi)*sp.cos(om)
y = R*sp.sin(fi)*sp.cos(om)
z = R*sp.sin(om)
```

```python
canon = x**2+y**2+z**2-R**2
sp.simplify(canon)
```

```python
List = plt.figure(figsize=(5,5))
ax = plt.axes(projection='3d')
coef = 2,4,8
fi = np.linspace(0,2*np.pi,200)
om = np.linspace(-2,2,10)
x = coef[0]*np.outer(np.cos(fi),np.cosh(om))
y = coef[1]*np.outer(np.sin(fi),np.cosh(om))
z = coef[2]*np.outer(np.ones_like(fi),np.sinh(om))
ax.plot_wireframe(x,y,z,rstride = 5,cstride = 5,alpha = 0.7)
ax.contourf(x,y,z,zdir='z',offset = np.min(z))
ax.contourf(x,y,z,zdir='y',offset = np.min(y))
ax.contourf(x,y,z,zdir='x',offset = np.min(x))
print('The equation of surface projection to XY:')
print('(x**2/2**2)+(y**2/4**2)-C=1 , where C = const')
print('The equation of surface projection to XZ:')
print('(x**2/2**2)-(z**2/8**2)+C=1 , where C = const')
print('The equation of surface projection to YZ:')
print("(y**2/4**2)-(z**2/8**2)+C=1 , where C = const")
```

```python
a,b,c = sp.symbols('a b c')
x = a*sp.cos(fi)*sp.cosh(om)
y = b*sp.sin(fi)*sp.cosh(om)
z = c*sp.sinh(om)
canon = (x**2/a**2)+(y**2/b**2)-(z**2/c**2)-1
sp.simplify(canon)
```

```python
Line = sp.Line(sp.Point(8,-1,-3),sp.Point(-4,3,3))
print('Parametric equations of a given line:')
T = Line.arbitrary_point()
sp.pprint(T)
x,z,t = sp.symbols('x z t')
ur = x**2-2*x-2*z
intersection = ur.subs([(x,T[0]),(z,T[2])])
print('t values:')
points = sp.solve(intersection)
sp.pprint(points)
```

```python
t = points[0]
M1 = [8 - 12*t, 4*t - 1, 6*t - 3]
t = points[1]
M2 = [8 - 12*t, 4*t - 1, 6*t - 3]
print('The intersection points of this line and the surface:')
print(M1,M2)
```
