```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure()
plt.axhline(y=0)
plt.axvline(x=0)
x = np.linspace(0.1,10,20)
y = x**2*np.log(x)**2
plt.xlabel('координата x')
plt.ylabel('координата y')
plt.title('Функция $y=x^{2}ln^{2}(x)$')
plt.grid(True)
plt.plot(x,y,'-ob')
plt.figure()
plt.axhline(y=0)
plt.axvline(x=0)
x = np.linspace(0.1,10,20)
y = np.log(x+np.sqrt(x**2+1))
plt.grid(True)
plt.xlabel('координата x')
plt.ylabel('координата y')
plt.title('Функция $y=ln(x+\sqrt{x^{2}+1})$')
plt.plot(x,y,':dr')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-2.0*np.pi,2.0*np.pi,20)
y1=np.sin(x)
gr1 = plt.plot(x,y1,':.g', label='y=sin(x)')
y2 = np.sin(x)-2
gr2 = plt.plot(x,y2,'-.py', label='y=sin(x)-2')
y3 = np.sin(x)+1
gr3 = plt.plot(x,y3,'--om', label='y=sin(x)+1')
plt.grid(True)
plt.xlabel('координата x')
plt.ylabel('координата y')
plt.axvline(x=0)
plt.axhline(y=0,color='g')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-2*np.pi,2*np.pi,40)
plt.plot(x,np.cos(x),':hc',label='y=cos(x)')
plt.plot(x,2*np.cos(x),'-xb',label='y=cos(x)')
plt.plot(x,0.3*np.cos(x),':^y',label='y=0.3cos(x)')
plt.plot(x,-np.cos(x),'--*m', label='y=-cos(x)')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.axvline(x=0)
plt.axhline(y=0,color='g')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-5,5,30)
plt.subplot(2,3,1)
plt.plot(x,np.abs(np.abs(x)-2))
plt.subplot(2,3,2)
plt.plot(x,np.abs(np.abs(x-2)-2))
plt.subplot(2,3,3)
plt.plot(x,np.abs(np.abs(x+2)-2))
plt.subplot(2,3,4)
plt.plot(x,np.abs(np.abs(2*x)-2))
plt.subplot(2,3,5)
plt.plot(x,np.abs(np.abs(0.5*x)-2))
plt.subplot(2,3,6)
plt.plot(x,np.abs(np.abs(-x)-2))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,np.pi,40)
y1 = np.cos(x)
y2 = -np.cos(x)
plt.title('cos(x)      -cos(x)')
plt.plot(x,y1,':g')
plt.plot(x,y2,'-.r')
plt.axvline(x=0, color='k')
plt.axhline(y=0, color='b')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure()
x = np.linspace(-3,100,20)
y = np.sqrt(x+3)
plt.plot(x,y,'-.dm')
plt.title('$\sqrt{x+3}$')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.figure()
x = np.linspace(-10,10,70)
y = np.sign(x)
plt.plot(x,y,'-*y')
plt.title('sing(x)')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.subplot(1,2,1)
x = np.linspace(-3,3,5)
plt.plot(x,x,'-r')
x = np.linspace(-3,3,20)
plt.plot(x,1/x,':m')
x = np.linspace(0.1,3,10)
plt.plot(x,1/np.sqrt(x),'-.k')
plt.subplot(1,2,2)
x = np.linspace(0,10,20)
plt.loglog(x,x,'--g')
x = np.linspace(1,10,20)
plt.loglog(x,1/x,'-.y')
x = np.linspace(1,10,20)
plt.loglog(x,1/np.sqrt(x),':k')
plt.show()
```
