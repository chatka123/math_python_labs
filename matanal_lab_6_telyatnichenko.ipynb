```python
import sympy as sp
from sympy.abc import x
y1 = (2*x-1)/(x-1)
y2 = (2*x-1)/(sp.Abs(x-1))
sp.plot(y1,(x,100,1000))
print('при x -> +inf, lim=2')
```

```python
sp.plot(y1,(x,-1000,-100))
print('при x -> -inf, lim=2')
```

```python
sp.plot((y1,(x,-10,0.8)),(y1,(x,1.2,10)))
print('при x -> inf, lim=2')
```

```python
sp.plot(y1,(x,1.0001,1.01))
print('при x -> 1+0, lim=inf')
print('при x -> 1-0, lim не существует')
print('при x -> 1, lim=inf')
```

```python
sp.plot(y1,(x,1.8,2.2))
print('при x -> 2+0, lim=3')
print('при x -> 2-0, lim=3')
print('при x -> 2, lim=3')
```

```python
ya = (1+x)**1/x
yb = (sp.sin(x**2/(sp.sqrt(1-x**3))))/(x*sp.ln(1-x**3))
sp.plot(ya,(x,1,100))
sp.plot(yb,(x,-1,1))
```

```python
A = sp.limit(ya,x,sp.oo)
B = sp.limit(yb,x,0)
sp.pprint(A)
sp.pprint(B)
```

```python
ya = sp.ln(x*sp.sin(x))
yb = (2**x)/(sp.factorial(x))
sp.plot(ya,(x,-10,10))
sp.plot(yb,(x,0,10))
```
