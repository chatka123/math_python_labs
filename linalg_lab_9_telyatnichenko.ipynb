```python
def gram_shmidt(*vectors):
    n = len(vectors)
    e1 = vectors[0]
    gs_vectors = list()
    gs_vectors.append(e1.astype(float))
    for i in range(1, n):
        e_i = vectors[i].astype(float)
        for j in range(0, i):
            e_i -= (np.dot(gs_vectors[j], vectors[i])/np.dot(gs_vectors[j], gs_vectors[j]))*gs_vectors[j]
            
        gs_vectors.append(e_i)
    return gs_vectors
```

```python
f1 = np.array([1, 1])
f2 = np.array([0, 1])
z = gram_shmidt(f1, f2)
```
