```python
def matrix(n,c,d):
    import numpy as np
    n = int(n)
    A = np.zeros((n,n))
    q = 0
    while q<n:
        A[q,q]=c
        A[q,q-1]=d
        A[q-1,q]=d
        q+=1
    A[0,-1]=0
    A[-1,0]=0
    print(A)
print('Enter the values for your matrix:')
n = input()
c = input()
d = input()
matrix(n,c,d)
```
