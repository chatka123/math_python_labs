```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
def vekt(m,q,w):
    p = np.array([m,q,w])
    return p
a=vekt(1,-2,0)
b=vekt(0,1,1)
c=vekt(1,2,2)
z=vekt(0,0,0)
X= Matrix([[1,0,1],[-2,1,2],[0,1,2]])
det = det(X)
if det>0:
    print('Определитель больше нуля, векторы линейно независимы и образуют базис, определитель равен', det)
elif det<0:
    print('Определитель меньше нуля, векторы линейно независимы и образуют базис, определитель равен', det)
else:
    print('Определитель равен нулю, векторы линейно зависимы и не образуют базис, определитель равен', det)
ax = plt.axes(projection='3d')
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
ax.set_zlim([-2,2])
ax.quiver(*z, *a, color='Purple')
ax.quiver(*z, *b, color='Black')
ax.quiver(*z, *c, color='Green')
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
def koord(x,y):
    r = np.array([x,y])
    return r
A=koord(-2,0)
B=koord(3,4)
C=koord(4,-1)
def vektr(N,K):
    t = K-N
    return t
AB=vektr(A,B)
BC=vektr(B,C)
CA=vektr(C,A)
def smart(one,two):
    cos = np.dot(one,two)/(np.linalg.norm(one)*np.linalg.norm(two))
    ugol = np.arccos(cos)
    ugol = 180 - math.degrees(ugol)
    return ugol
bac=smart(AB,CA)
abc=smart(AB,BC)
bca=smart(BC,CA)
print("Угол BAC треугольника равен",bac,'градусов')
print("Угол ABC треугольника равен",abc,'градусов')
print("Угол BCA треугольника равен",bca,'градусов')
plt.arrow(-2,0,AB[0],AB[1], width=0.05, length_includes_head=True,facecolor='Red')
plt.arrow(3,4,BC[0],BC[1], width=0.05, length_includes_head=True,facecolor='Black')
plt.arrow(4,-1,CA[0],CA[1], width=0.05, length_includes_head=True,facecolor='Green')
plt.show()
sum=bac+abc+bca
if sum>180:
    print('Сумма углов треугольника больше 180 градусов,переделывай')
elif sum<180:
    print('Сумма углов треугольника меньше 180 градусов,переделывай')
else:
    print('Сумма углов треугольника равна 180 градусам, все правильно')
print(sum)
```

```python
import numpy as np
def koord(x,y,z):
    r = np.array([x,y,z]) #создает массив с координатами точки
    return r
def vektr(one,two):
    q = np.array([two[0]-one[0],two[1]-one[1],two[2]-one[2]]) #вычисляет координаты вектора и заносит в массив
    return q
def dlina(r):
    t = np.sqrt(r[0]**2+r[1]**2+r[2]**2) #считает модуль вектора 
    return t

A=koord(1,3,-1)
B=koord(2,-1,4)
C=koord(5,0,3)
AB=vektr(A,B)
BC=vektr(B,C)
CA=vektr(C,A)
modAB=dlina(AB) #длинна вектора AB
modBC=dlina(BC) #длинна вектора BC
modCA=dlina(CA) #длинна вектора CA
p=(modAB+modBC+modCA)/2 #полупериметр сторон треугольника
ploshad=np.sqrt(p*(p-modAB)*(p-modBC)*(p-modCA)) #площадь треугольника
Q=dlina(np.cross(AB,BC)) #модуль векторного произведения векторов AB и BC
print('Площадь треугольника и векторное произведение его векторов различаются в',Q/ploshad,'раз')
```

```python
import numpy as np
from sympy import *
def vektr(x,y,z):
    r = np.array([x,y,z]) #функция задаёт координаты вектору в виде массива
    return r
AB=vektr(3,4,0) 
AC=vektr(-3,0,1) #задание координат векторам
AD=vektr(0,2,5) 
vek_proizv=np.cross(AB,AC) #вектрное произведение векторов AB и AC
skl_proizv=np.dot(vek_proizv,AD) #скалярное произведение векторного произведения выше и вектора AD
smesh_proizv=skl_proizv #объявление переменной
V = 1/6*smesh_proizv #подсчет объема 
print('Объем искомой пирамиды равен',V)
w = Matrix([AB,AC,AD]) #начало проверки (задачия координат векторов в матрицу)
i = det(w) #подсчет определителя
if i > smesh_proizv:
    print('Ошибка, смешанное произведение меньше определителя матрицы, это невозможно')
elif i < smesh_proizv:
    print('Ошибка, смешанное произведение больше определителя матрицы, это невозможно') #некоторы цикл для автоматизации сравнения результата
else:
    print('Все правильно, смешанное произведение равное',smesh_proizv,'равно определителю матрицы, равному',i)  
```
