```python
import numpy as np
import sympy as sp
```

```python
def jointness(a,b,variable_n):    
    x,y,z,q = sp.symbols('x y z q')
    A = a
    B = b
    AIB = A.row_join(B.T)
    variable = variable_n
    print(f'A rang :{A.rank()}')
    print(f'AIB rang:{AIB.rank()}')
    if (A.rank()==variable and AIB.rank()==variable) or A.rank()==AIB.rank() and AIB.rank()<variable:
        sol = sp.linsolve((A, B),[x,y,z,q])
        print(f'Solution of this system:{sol}')
    else:
        print('This system is non-local')
```

```python
def fundamental_system(a,n):
    C1, C2, C3, C4, C5, C6 = sp.symbols("C1 C2 C3 C4 C5 C6")
    AA = a
    print('Rang of system =', AA.rank())
    E=AA.nullspace()
    print(f'solutions of fundamental system:{E}')
    c = [C1, C2, C3, C4, C5, C6]
    answer = sp.Matrix([[0]]*n)
    for i in range(len(E)):
        answer = answer+c[i]*E[i]   
    print(f'General solution:{answer}')
```

```python
fundamental_system(sp.Matrix([[2, -3, 1], [1, 1, 1], [3, -2, 2]]),3)
```

```python
fundamental_system(sp.Matrix([[1,2,4,-3],[3,5,6,-1],[4,4,-2,3], [3,8,24,-19]]),4)
```
