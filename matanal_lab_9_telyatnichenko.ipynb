```python
import numpy as np
poly = [1,0,0,-3.2,0,3,0,3]
points = [-1,4,2.2,np.pi]
values = np.array([])
for i in points:
    x = np.polyval(poly,i)
    values = np.append(values, x)
    print('the value of polynomial at x =',i,':',round(x,4))
print('the values of polynomial at all point:',values)
```

```python
poly = [2,0,0,-3,0,0,7,0,-2]
x = np.roots(poly)
print('all values of desired polynomial:',x)
for i in x:
    print('the value of polynomial at x =',f'{i:.4f}')
    print(np.round(np.polyval(poly,i)))
```

```python
poly1 = [1,2,0,0,-3,0,4]
poly2 = [1,0,-3,1]
umn = np.convolve(poly1,poly2)
chast = np.polydiv(poly1,poly2)
print(umn)
print(chast)
```

```python
def sum_polynomial(polyONE,polyTWO):
    poly1 = polyONE
    poly2 = polyTWO
    len1 = len(poly1)
    len2 = len(poly2)
    if len1>len2:
        support = np.zeros(len1-len2)
        poly2 = np.append(support,poly2)
        print(poly1+poly2)
    elif len2>len1:
        support = np.zeros(len2-len1)
        poly1 = np.append(support,poly1)
        print(poly1+poly2)
    else:
        print(poly1+poly2)
```

```python
sum_polynomial([1,2,0,0,-3,0,4],[1,0,-3,1])
```

```python
poly1 = [1,-1,0,-3,0,-2]
poly2 = [1,0,0,0,0,-3,1]
x = np.polydiv(poly1,poly2)
print('derivative of p(x):',np.polyder(poly1))
print('derivative of q(x):',np.polyder(poly2))
print('derivative of p(x)*q(x):',np.polyder(np.convolve(poly1,poly2)))
print('derivative of p(x)/q(x):',np.polyder(x[1]))
```

```python
def derivative_n(pol,n,x0):
    poly = pol
    derivative = np.polyder(poly,n)
    print('polynomial derivative of n-methods:',derivative)
    value = np.polyval(derivative,x0)
    print('polynomial value of the derivative of n-methods:',value)
```

```python
derivative_n([1,0,0,-1,0,0,0,0,3,0,-2],5,0.2)
```

```python
def factorial(n):
    value = 1
    while n>1:
        value = value * n
        n -= 1
    return value
```

```python
factorial(6)
```

```python
def b_coef(Poly,x0):
    poly = Poly
    bk = []
    for i in range(len(poly)-1,-1,-1):
        derivative = np.polyder(poly,i)
        print(np.poly1d(derivative))
        b = (np.polyval(derivative,x0))/(factorial(i))
        bk.append(b)   
    return bk
```
