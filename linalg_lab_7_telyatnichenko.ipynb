```python
def full_square(A=0,B=0,D=0,E=0,F=0):
    x,y = sp.symbols('x y')
    if A!=0 and D!=0:
        resx = (sp.sqrt(A)*x+D/(2*sp.sqrt(A)))**2-(D/(2*sp.sqrt(A)))**2
    else:
        resx = A*x**2+D*x
    if B!=0 and E!=0:
        resy = (sp.sqrt(B)*y+E/(2*sp.sqrt(B)))**2-(E/(2*sp.sqrt(B)))**2+F
    else:
        resy = B*y**2+E*y+F
    otv = resx+resy
    return otv
kriv = full_square(4,9,-8,-36,4)
print('The desired equation of the curve:')
sp.pprint(kriv)
print('Focal points of the desired curve:',kriv)
```
