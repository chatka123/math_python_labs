```python
def funkcia(x1,y1,z1,q1,q2,q3,x2,y2,z2,x3,y3,z3):
    A = Point(x1,y1,z1)
    P = Plane((A),normal_vector=(q1,q2,q3))
    C = Point(x2,y2,z2)
    D = Point(x3,y3,z3)
    L2 = Line(Point(x2,y2,z2),Point(x3,y3,z3))
    CP = P.projection(C)
    DP = P.projection(D)
    if CP==DP:
        print(' ,   A  L2:')
        print(CP.distance(A))
    else:
        print(' ,     PL2:')
        PL2 = Line(CP,DP)
        print(A.distance(PL2))
funkcia(2,-1,0,3,4,2,7,1,3,10,5,5)
```
