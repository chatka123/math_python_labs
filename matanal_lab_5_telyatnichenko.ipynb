```python
import sympy as sp
import math as mh
import numpy as np
import matplotlib.pyplot as plt
from sympy.abc import a,n,x
#a
b = sp.solve_univariate_inequality(n>0.1,n,relational=False)
LIMIT = sp.limit((n**3-1)/(2*n**3+2*n),n,mh.inf)
n01 = sp.solve_univariate_inequality(sp.Abs(((n**3-1)/(2*n**3+2*n))-LIMIT)<0.01,n,relational=False)
n01 = sp.Intersection(n01,b)
n02 = sp.solve_univariate_inequality(sp.Abs(((n**3-1)/(2*n**3+2*n))-LIMIT)<0.001,n,relational=False)
n02 = sp.Intersection(n02,b)
print('      0.01, n0:')
sp.pprint(n01)
print('      0.001, n0:')
sp.pprint(n02)
def smart(xn,n0,a,eps,x1,x2,y1,y2):
    reshenia = np.linspace(n0 - 2,n0 + 10,13)
    plt.plot(reshenia, xn(reshenia),'-mo')
    plt.axhline(a - eps,color='r')
    plt.axhline(a + eps,color='k')
    plt.axis([x1,x2,y1,y2])
    plt.grid()
    plt.show()
smart(lambda n: ((n**3-1)/(2*n**3+2*n)),8,LIMIT,0.01,6.5,18.5,0.475,0.511)
smart(lambda n: ((n**3-1)/(2*n**3+2*n)),23,LIMIT,0.001,21.5,33.5,0.4985,0.5015)
```
