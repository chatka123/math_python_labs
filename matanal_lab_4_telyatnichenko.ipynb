```python
import numpy as np
a=pow(complex(256,0),1/8)
b=np.abs(a)
c=np.angle(a)
d=np.arange(0,8)
x=b**(1/8)*(np.cos((c+2*np.pi*d)/8)+1j*np.sin((c+2*np.pi*d)/8))
print('___________________________________________________________________________________________________________________________________')
for i in x:
    if i >0:
        print(':',i,'   ')
    elif i<0:
        print(':',i,'   ')
print('___________________________________________________________________________________________________________________________________')
X = np.imag(x)
for a in X:
    r = round(a)
    if r > 0:
        print('   ',a,'   Re')
    elif r < 0:
        print('   ',a,'   Re')
    else:
        print('   ',a,'   Re')
print('___________________________________________________________________________________________________________________________________')
Y = np.angle(x)
for i in Y:
    if i > -np.pi/3:
        print('  ',i,'   ,  -pi/3')
    else:
        print('  ',i,'  ')
prov = np.array([0,0,0,0,0,0,0,0])
otv = np.isclose(X,prov)
print(otv)
```

```python
import numpy as np
import numpy.random as rng
import matplotlib.pyplot as plt
x = 3 - 6*rng.rand(5500)
y = 3 - 6*rng.rand(5500)
z = x + 1j*y
L = (np.abs(z-1+1j)>=1)&(np.abs(z-1+1j)<=2)
plt.plot(x[L],y[L],'y*')
plt.grid(True)
plt.axis("equal")
plt.title('')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```

```python
import numpy as np
z = pow(complex(-256,0),1/8)
arg = np.angle(z)
mod = np.abs(z)
mass = np.arange(0,8)
x = mod**(1/8)*(np.cos((arg+2*np.pi*mass)/8)+1j*np.sin((arg+2*np.pi*mass)/8))
print('_____________________________________________________________________________________________')
for i in x:
    if i < 0:
        print('',i,'   ')
print('_____________________________________________________________________________________________')
realX = np.imag(x)
for I in realX:
    X = round(I)
    if X>0:
        print('    ',I,'   ')
    elif X<0:
         print('    ',I,'   ')
    else:
         print('    ',I,'   ')
print('_____________________________________________________________________________________________')
Y = np.angle(x)
for i in Y:
    if i > 2*np.pi/3:
        print('  ',i,'   ,  2pi/3')
    else:
        print('  ',i,'  ')
```

```python
import numpy as np
import numpy.random as rng
import matplotlib.pyplot as plt
x = 2-4*rng.rand(37000)
y = 2-4*rng.rand(37000)
z = x+1j*y
arg = np.abs(np.angle(z))
modl = np.abs(z-1)
modl2 = np.sqrt((np.real(z)-0.3)**2+(np.imag(z)-1.3)**2)
resh = (arg>=np.pi/4)&(modl<=2)&(modl2>=0.3)
plt.plot(x[resh],y[resh],'dm')
plt.axis('equal')
plt.grid()
plt.title('')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.show()
```
