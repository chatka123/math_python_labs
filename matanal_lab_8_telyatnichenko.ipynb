```python
def derivative(y,x0):
    import sympy as sp
    output = sp.limit((y(x0+dx)-y(x0))/dx,dx,0)
    print(round(output,3))
    return round(output,3)
```

```python
oneA = derivative(lambda x:x**1/3,1)
twoA = derivative(lambda x:x**1/3,2)
threeA = derivative(lambda x:x**1/3,-3)
```

```python
import sympy as sp
x = sp.symbols('x')
f = sp.atan(sp.sqrt(x))**2
y = sp.diff(f,x,1)
print('function derivative:')
sp.pprint(y)
print('function value at x0=0,5:',y.subs(x,0.5))
```

```python
f = 3**sp.asin(x)**2
y = sp.diff(f,x,1)
print('function derivative:')
sp.pprint(y)
print('function value at x0=0,5:',sp.N(y.subs(x,0.5),4))
```

```python
def tangent_line(expression,x0):
    x = sp.symbols('x')
    f = expression
    y = sp.diff(f,x,1)
    f0 = f.subs(x,x0)
    y0 = y.subs(x,x0)
    Y = f0+y0*(x-x0)
    sp.plot((f,(x,x0-1,x0+1)),(Y,(x,x0-1,x0+1)))
```

```python
tangent_line(sp.cos(3*x)**3,sp.pi/4)
```

```python
tangent_line(sp.E**(2*x),1)
```
