```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt #     
L1 = Line((0,0),(2,3)) #   L1
L2 = L1.parallel_line(Point(-2,2)) #      L2    sympy ".parallel_line()"    .
L3 = L1.perpendicular_line(Point(-2,2)) #---//---    
urav1 = L1.equation() 
urav2 = L2.equation()
urav3 = L3.equation() #      sympy ".equation()"
print('  :',urav1)
print('  :',urav2)
print('  :',urav3) #  
x = np.linspace(-20,20,100) #   x
y1 = -3*x/-2 
y2 = (-3*x-10)/-2
y3 = (-2*x+2)/3 #      
plt.plot(x,y1,'m',label='L1')
plt.plot(x,y2,'g',label='L2')
plt.plot(x,y3,'r',label='L3') #     x  y,      
plt.legend() #  
plt.grid() # 
plt.xlabel('x') 
plt.ylabel('y') #  
plt.title('') # 
plt.axis('equal') #   
plt.axhline(y=0,color='black')
plt.axvline(x=0,color='black') # 
plt.show() # 
print('') 
s1=L1.angle_between(L3)
print('   L3  L1 :') #         ".angle_between()"    
s1                                         #  L1  L3  90 
```
