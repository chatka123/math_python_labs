```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt #ввожу нужные для зажания библиотеки 
L1 = Line((0,0),(2,3)) #задаю точки прямой L1
L2 = L1.parallel_line(Point(-2,2)) #нахожу точки через которые проходит прямая L2 с помощью функции sympy ".parallel_line()" исходя из условия паралл.
L3 = L1.perpendicular_line(Point(-2,2)) #---//--- исходя из условия перпендикулярности
urav1 = L1.equation() 
urav2 = L2.equation()
urav3 = L3.equation() #нахожу уравнения прямых с помощью функции sympy ".equation()"
print('уравнение исходной прямой:',urav1)
print('уравнение параллельной прямой:',urav2)
print('уравнение перпендикулярной прямой:',urav3) #вывожу их 
x = np.linspace(-20,20,100) #генерирую последовательность чисел x
y1 = -3*x/-2 
y2 = (-3*x-10)/-2
y3 = (-2*x+2)/3 #задаю уранения всех трех функции соответственно 
plt.plot(x,y1,'m',label='L1')
plt.plot(x,y2,'g',label='L2')
plt.plot(x,y3,'r',label='L3') #строю три графика на плоскости x и y, а так же даю им имена
plt.legend() #вывожу легенду карты
plt.grid() #наношу сетку
plt.xlabel('x') 
plt.ylabel('y') #даю имена осям
plt.title('Прямые') #имя графика
plt.axis('equal') #отформатировал размеры графического окна
plt.axhline(y=0,color='black')
plt.axvline(x=0,color='black') #наношу оси
plt.show() #вывожу результат
print('ПРОВЕРКА') 
s1=L1.angle_between(L3)
print('Угол между прямыми L3 и L1 равен:') #провожу проверку с помощью функции подсчета угла между прямыми ".angle_between()" и удостовериваюсь что 
s1                                         #угол между L1 и L3 равен 90 градусов
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import * #ввел библиотеки 
A = Point(2,-2)
B = Point(3,5)
C = Point(6,1) #ввел точки 
AB = Line((A),(B))
BC = Line ((B),(C))
CA = Line ((C),(A)) #задал прямые треугольника 
urAB = AB.equation()
urBC = BC.equation()
urCA = CA.equation() #нашел уравнения каждой прямой треугольника
print('Уравнение прямой, содержащей AB:',urAB)
print('Уравнение прямой, содержащей BC:',urBC)
print('Уравнение прямой, содержащей CA:',urCA) #вывел их 
x1 = np.linspace(2,3,10)
y1 = 7*x1-16
plt.plot(x1,y1,color='g') #задал и построил отрезок AB треугольника
x2 = np.linspace(3,6,10)
y2 = -4*x2/3+9
plt.plot(x2,y2,color='g') # ---//--- BC ---//---
x3 = np.linspace(2,6,10)
y3 = 3*x3/4-3.5
plt.plot(x3,y3,color='g') #----//--- CA ---//---
H = CA.perpendicular_line(Point(3,5)) #нашел точки через которые проходит высота из вершины B треугольника пользуясь её свойствами перпендикулярности
urH = H.equation() #нашел уравнение прямой содержащей высоту (точки высоты)
print('Уравнение прямой, содержащей высоту треугольника H проведенную из B:',urH)
print('Уравнения прямых BC:',urBC,'и H:',urH,'совпадают, треугольник прямоугольный')
Hvek = np.array([[1,7]]) #задал с помощью массива координаты вектора высоты совпадающие с координатами вектора BC
Hdl = np.sqrt(1**2+7**2) #нашел длинну высоты H (BC)
print('Длинна высоты треугольника H:',Hdl)
plt.plot(A[0],A[1],'or') 
plt.text(A[0]-0.4,A[1],'A')
plt.plot(B[0],B[1],'or')
plt.text(B[0]+0.3,B[1],'B')
plt.plot(C[0],C[1],'or')
plt.text(C[0]+0.3,C[1],'C') #построил точки и дал им названия 
plt.axis('equal') #отформатировал размеры графического окна 
M = Point(4.5,3) #ввел точку M середину BC
AM = Line((A),(M)) #задал прямую AM через две точки, она и является медианой треугольника из вершины A
urAM = AM.equation() #нашел уравнение прямой, содержащей медиану AM
print('Уравнение прямой, содержащей медиану треугольника AM проведенную из A:',urAM)
M = np.linspace(2,4.5,10) 
yM = 2*M-6
plt.plot(M,yM,color='y')
plt.plot(4.5,3,'or')
plt.text(4.7,3,'M') #задал, построил медиану AM  и точку M а так же подписал её в граф. окне
AMdl = np.sqrt(2.5**2+5**2) #нашел длинну медианы AM
print('Длинна медианы треугольника AM:',AMdl)
s1=AB.smallest_angle_between(BC) #наъожу угол между AB и BC он оказывается равен 45 градусам, азначит треугольник равнобедренный
print('Угол между AB и BC равен ',s1,',треугольник р/б')  
T = Point(2.5,1.5) #ввел точку T, середину гипотенузы AB, чтобы биссектрисой (медианой) из вершины прямого угла р/б треугольника  был отрезок CT 
T = Line((C),(T)) 
urT = T.equation() #нахожу уравнение биссектрисы 
print('Уравнение биссектрисы треугольника проведенной из C:',urT)
T = np.linspace(6,2.5,100)
yT = -T/7+13/7
plt.plot(T,yT,color='m')
plt.plot(2.5,1.5,'or')
plt.text(2,1.5,'T') #задал, построил биссектрису CT и точку T а так же подписал её в граф. окне
plt.grid()
plt.show()
```
