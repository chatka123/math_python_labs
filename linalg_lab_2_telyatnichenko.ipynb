```python
import numpy as np 
A = np.matrix('3 -5;2 7')
B = np.matrix('13;81')
X = np.matrix(A)
X[:,0] = B
Y = np.matrix(A)
Y[:,1] = B
det_A = np.linalg.det(A)
x = np.linalg.det(X) / det_A
print(round(x))
y = np.linalg.det(Y) / det_A
print(round(y))
p = np.array([[round(x)],[round(y)]])
number = A@p - B
for i in number:
    if i>0:
        print('Переделывай')
    elif i <0:
        print('Переделывай')
    else:
        print('Все сошлось')
print(number)
```

```python
#(*)Выводы: в этой части упражнения я нашел определитель второго порядка для данной матрицы, так как численный результат корней системы получился с так называемой "плавающей точкой" я 
#использовалбазовую функцию pyhton "round()" и провел округление до целых с помощью неё. В конце после проведения проверки с помощью заданного в упражнении равенства ответы сошлись. Так
#кроме того для качественного оформления задачи я использовал вшитые в питон операторы for in if else для создания цикла и проверки результатов.Кроме того важно уточнить, что для проверки корней была создана матрица типа 'вектор-столбец'
#для возможности перемножения матриц изходя из равенства для проверки решения 
```

```python
from sympy import *
A = Matrix([[3,-5],[2,7]])
B = Matrix([[13],[81]])
X = Matrix(A)
X[:,0] = B
Y = Matrix(A)
Y[:,1] = B
x = det(X) / det(A)
y = det(Y) / det(A)
print(x)
print(y)
p = Matrix([[x],[y]])
number = A@p - B
for i in number:
    if i>0:
        print('Переделывай')
    elif i <0:
        print('Переделывай')
    else:
        print('Все сошлось')
number
```
