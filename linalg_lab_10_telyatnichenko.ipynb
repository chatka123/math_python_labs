```python
def full_square(A=0,B=0,D=0,E=0,F=0):
    x,y = symbols('x y')
    if A!=0 and D!=0:
        resx = (sqrt(A)*x+D/(2*sqrt(A)))**2-(D/(2*sqrt(A)))**2
    else:
        resx = A*x**2+D*x
    if B!=0 and E!=0:
        resy = (sqrt(B)*y+E/(2*sqrt(B)))**2-(E/(2*sqrt(B)))**2+F
    else:
        resy = B*y**2+E*y+F
    otv = resx+resy
    return otv
```

```python
def simple_square(a, b, d, e, f):
    if a>0:
        f_x = (d/2/sqrt(a))**2
        x_coef = [sqrt(a), sign(d)*sqrt(f_x), 1]
    elif a<0:
        f_x = -abs((d/2/sqrt(-a))**2)
        x_coef = [sqrt(-a), sign(d)*sqrt(-f_x), -1]
    else:
        f_x = 0
        x_coef = [0, d, 1]

    if b>0:
        f_y = (e/2/sqrt(b))**2
        y_coef = [sqrt(b), sign(e)*sqrt(f_y), 1]
    elif b<0:
        f_y = -abs((e/2/sqrt(-b))**2)
        y_coef = [sqrt(-b), sign(e)*sqrt(-f_y), -1]
    else:
        f_y = 0
        y_coef = [0, e, 1]
    f_fin = f-f_x-f_y
    return x_coef, y_coef, f_fin

```

```python
x_coef, y_coef, f_fin = simple_square(5, 30, 14, -12, 9)
x, y = symbols('x1 y1')
if x_coef[0]!=0:
    xx = x_coef[2]*(x_coef[0]*x+x_coef[1])**2
else:
    xx = x_coef[1]*x
if y_coef[0]!=0:
    yy = y_coef[2]*(y_coef[0]*y+y_coef[1])**2
else:
    yy = y_coef[1]*y
F1 = xx+yy+f_fin
print(f"В новой системе координат y2={y_coef[0]*y+y_coef[1]}")
print(f"В новой системе координат x2={x_coef[0]*x+x_coef[1]}")
print(f"скаляр при x^2: {x_coef[2]}")
print(f"скаляр при y^2: {y_coef[2]}")
print(f"свободный член: {f_fin}")
```
